# Adding details to a graph

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot (x, y, color="red", label="x^2", marker='^', markersize=10.2, markeredgecolor="green", 
          linewidth=1.5, linestyle="dotted")
#Colors can also be passed with their initials (some) or as hexadecimal codes
#Default markersize=6, linewidth=2 (decimal sizes are also accepted)
#Marker edge color just outlines the markers with the specified color

#Marker options: '.', 'o'(like bigger '.'s), 'x',  'X'(bold), '+', '1', '2', '3', '4', ',' '' and ''''''(nothing) (default)
#'s'(square), 'd'(diamond), 'v'(inverted triangle), '8'(same as 'o'), '^'(triangle), '*' (star)
#Linestyle options: 'dashed' (or --), 'dotted', 'solid' (or -) (default)


plt.title('Square graph', fontdict={'fontname':'Chiller', 'size':20})
plt.xlabel('Numbers', fontdict={'fontname':'Comic Sans MS', 'size':13})
plt.ylabel('Squares', fontdict={'fontname':'Comic Sans MS', 'size':13})

#For a range of 1-5 (or 0-4 and so on), it displays the numbers as float values on both the axes. 
plt.xticks([1, 2, 3, 4, 5])
plt.yticks(y)    #Inputting a list like this works, too
#This helps specify the exact points we want to see on the graph 
#(can include any number of points we want to see in any range (only for line graphs, though))

plt.legend()       #Puts a legend at the top right corner by default (inside the graph)
plt.figlegend()    #Puts a legend at the top right corner by default (outside the graph)
#Both the above lines gave a warning 'No handles with labels found to put in legend.' 
#but worked after I added the label parameter in plt.plot

plt.show()
```

```python
#The "Shorthand" Notation
#Pass the color, marker and linestyle as one parameter in their short forms (short forms are a must)
#However, x-- (x is any color) doesn't work, as the --- confuses the program

a = [4, 8, 12, 16, 20]
b = [2, 4, 6, 8, 10]

plt.plot(b, a, 'b*--', markersize=15, markeredgecolor="pink")
plt.show()
```

# Bar Graph

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dragonfruit']
values = [15,  12, 14, 13]

plt.figure (figsize=(5, 3), dpi=100)        #This line should go below the plot/bar line to affect it
bg = plt.bar (labels, values, color='g', label='Fruits')    #type(bg) = matplotlib.figure.Figure

bg[0].set_hatch("O")     #Gives a pattern to the bar
bg[1].set_hatch("\\")    #Had to write \ twice to avoid "scanning end of literal"
bg[2].set_hatch(".")
bg[3].set_hatch("*")
#Even if I pass something random as the argument, I won't get an error (as long as there is only one and within '' or "")

#plt.figlegend(labels) not helpful in this case (could be mixed with plt.legend(), same outcome)
plt.show()
```

```python
#Shortening the Set-Hatch process

labels = ['Apples', 'Bananas', 'Cherries', 'Dragonfruit']
values = [15,  12, 14, 13]

plt.figure (figsize=(5, 3), dpi=100)        #This line should go below the plot/bar line to affect it
bg = plt.bar (labels, values, color='orange', label='Fruits')    #type(bg) = matplotlib.figure.Figure
plt.title("Fruits \n", fontdict={'color':'brown', 'size':15, 'fontweight':'bold'})    #\n adds line break

hatches = ['|', "X+X+X+X", "//\\", "OO"]

for i in bg:
    #i.set_hatch(hatches.pop())     #Assigns patterns starting from the end
    i.set_hatch(hatches.pop(0))     #Assigns patterns starting from the beginning
    
plt.savefig('Cool Chart.png', dpi=300)
plt.show()
```

# A more detailed pie chart

```python
fifa.Weight
#Here, it has values with 'lbs', meaning that the values are being treated as strings
```

```python
#So, to remove the 'lbs'

#fifa.Weight = [x.strip('lbs') for x in fifa.Weight] gave AttributeError as 'lbs' may not be there in every row
#fifa.Weight = [x.strip('lbs') if type(x)==str for x in fifa.Weight] gave SyntaxError without the else clause
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
#Converting into int as stripping will not change its initial datatype (str)

fifa.Weight
#Now it works
```

```python
#Creating weight categories
light = fifa.loc[fifa.Weight<=130].count()[0]    #Take only the first value in the series of counts
med_light = fifa.loc[(fifa.Weight>130) & (fifa.Weight<=140)].count()[0]
medium = fifa.loc[(fifa.Weight>140) & (fifa.Weight<170)].count()[0]
med_heavy = fifa.loc[(fifa.Weight>=170) & (fifa.Weight<190)].count()[0]
heavy = fifa.loc[fifa.Weight>=190].count()[0]

weights = np.array([light, med_light, medium, med_heavy, heavy])
labels = np.array(['Light', 'Medium-Light', 'Medium', 'Medium-Heavy', 'Heavy'])
#colours = np.array(['yellow', 'teal', 'green', 'cyan', 'navy']) (Checking out plt.style)
explode = [1, 0.5, 0, 0, 0.25]
#Moves the slices out (away from the center) by the factor specified

plt.style.use('ggplot')    #Applies some pre-defined colour (and maybe size) formatting to the chart

plt.pie(weights, labels=labels, autopct="%.1f%%", radius=1, pctdistance=0.8, explode=explode)
#The radius simply multiples the radius of the piechart by that factor (enlarging its area by a square factor)
#pctdistance = Percentage Distance (decides the ratio- distance of percentages from centre : radius)

plt.title("Player Distribution by Weight Categories ", fontdict={'fontname':'Calibri', 'size':20, 'color':'indigo'})
plt.figlegend()
plt.show()
```

# Box and Whiskers Charts

```python
#Let's plot one to see the spread and median of players in different clubs and their overall scores

print ([fifa['Club']])    #Let's pick Juventus, Manchester United and Tranmere Rovers

juv = fifa.loc[fifa['Club']=='Juventus']['Overall']
mu = fifa.loc[fifa['Club']=='Manchester United']['Overall']
tr = fifa.loc[fifa['Club']=='Tranmere Rovers']['Overall']
#Specifying which column we want to pick from the rows satisfying the specified condition(s)

print (juv)
print (mu)
print (tr)
```

```python
#Plotting the charts
plt.style.use('default') #Changing the style again to discontinue the previous one

#Remember to pass 'labels' as a list for box plots
plt.boxplot (juv, labels=["Juventus"])
plt.title ('Juventus Overall Scores')
plt.xlabel ('Team')
plt.ylabel ('Player Overall Scores')
plt.show()

#-------------------------------------------------------------

plt.boxplot (mu, labels=["Manchester United"])
plt.title ('Manchester United Overall Scores')
plt.xlabel ('Team')
plt.ylabel ('Player Overall Scores')
plt.show()

#--------------------------------------------------------

plt.boxplot (tr, labels=['Tranmere Rovers'], widths=0.1)    #Adjusts the width of the box (along the X-Axis)
plt.title ('Tranmere Rovers Overall Scores')
plt.xlabel ('Team')
plt.ylabel ('Player Overall Scores')
plt.show()

#The circles on the graphs represent outliers
```

```python
#Box charts also allow comparative charts (pass the series as elements of a list) and their colours are changed differently

label4 = ['Juventus', 'Manchester United', 'Tranmere Rovers']

plt.figure (figsize=(6, 6), dpi=150)    #(x, y)
boxes = plt.boxplot ([juv, mu, tr], labels=label4, patch_artist=True) #patch_artist=True allows changing box fill colour

for i in boxes['boxes']:    #Specifying 'plots' again, now as the index
    i.set(color="blue", linewidth=1)    #Not working for some reason (reason being that only 'boxes' works, not 'plots')
    i.set(facecolor="lavender")    #Changed box border colour, border width, fill colour

for i in boxes['whiskers']:    #'fliers', 'medians', 'caps' are other indices
    i.set(color="purple", linewidth=2)    
    
for i in boxes['medians']:    #Another way to do this is define the dictionary for medianprops in plt.boxplot
    i.set(color="indigo", linewidth=3) 
    
plt.title ('Juventus Overall Scores')
plt.xlabel ('Teams')
plt.ylabel ('Player Overall Scores')
plt.show()

#Removed the other comparison charts (comparing 2 clubs at a time) as the final one emcompasses all three
```
